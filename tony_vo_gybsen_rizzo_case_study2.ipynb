```python
def convert_money(value):
    if isinstance(value, str):
        value = (value.replace('$', ''). replace(',', ''))
        try:
            return float(value)
        except ValueError:
            return None
        return None

tsa2['Claim Amount'] = tsa2['Claim Amount'].apply(convert_money)
tsa2['Close Amount'] = tsa2['Close Amount'].apply(convert_money)
```
